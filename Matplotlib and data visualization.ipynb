```python
heart_df=pd.read_csv("heart-disease.csv")
```

```python
heart_df
```

```python
heart_df.plot.hist(figsize=(10,30),subplots=True);
```

```python
## which one to use pyplot or matplotlib oo method?
# when plotting something quickly it is ok to use pyplot method
#When plotting something more advanced use OO method
```

```python
heart_df.dtypes
```

```python
over_50=heart_df[heart_df["age"]>50]   ##for filtering a column in data frame make sure to use []
over_50
```

```python
## pyplot Method
over_50.plot(kind="scatter",x="age",y="chol",c="target",title="Cholesterol level for age above 50 years");
```

```python
# OO(object orented method) Method -this looks mostly like pyplot

fix,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
               x="age",y="chol",
               c="target",
               title="Cholesterol level for age above 50 years",
               ax=ax);
#ax.set_xlim([25,100]);
```

```python
#lets recreate OO method
fig,ax=plt.subplots(figsize=(10,6))
#plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])

## customize the plot
ax.set(title="Heart Disease and Cholestral",
      xlabel="Age",
      ylabel="Cholestral")
## add legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
#1.sub plots for age, chol, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10))
#2.add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])
##3. customize the plot ax0
ax0.set(title="Heart Disease and Cholestral",
      xlabel="Age",
      ylabel="Cholestral")
## 4.add legend for ax0
ax0.legend(*scatter.legend_elements(),title="Target");

#5. add a horizontal line for ax0
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");
#repeat the same process for ax1

#2.add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"])
##3. customize the plot ax1
ax1.set(title="Heart Disease and Cholestral",
      xlabel="Age",
      ylabel="Heart Rate")
##4. add legend for ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#5. add a horizontal line for ax1
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");
```

```python
# remove x axis from ax0 in order to have one plot alone
#1.sub plots for age, chol, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10))
#2.add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])
##3. customize the plot ax0
ax0.set(title="Heart Disease and Cholestral",
#xlabel="Age",
      ylabel="Cholestral")
## 4.add legend for ax0
ax0.legend(*scatter.legend_elements(),title="Target");

#5. add a horizontal line for ax0
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");
#repeat the same process for ax1

#2.add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"])
##3. customize the plot ax1
ax1.set(title="Heart Disease and Cholestral",
      xlabel="Age",
      ylabel="Heart Rate")
##4. add legend for ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#5. add a horizontal line for ax1
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

## Add a super title for our figure
fig.suptitle("Heart Disease study by Yatheesha",fontsize=16,fontweight="bold");
```

```python
## Lets st the style first
plt.style.use('seaborn-whitegrid')
#1.sub plots for age, chol, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10))
#2.add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"],
                   cmap="winter")

# change the x axis limits(do this step atlast)
ax0.set_xlim([50,80])


##3. customize the plot ax0
ax0.set(title="Heart Disease and Cholestral",
#xlabel="Age",
      ylabel="Cholestral")
## 4.add legend for ax0
ax0.legend(*scatter.legend_elements(),title="Target");

#5. add a horizontal line for ax0
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");
#repeat the same process for ax1

#2.add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"],
                   cmap="plasma")                            ##this changes the colour schema

# change the x axis limits #do this step atlast
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])


##3. customize the plot ax1
ax1.set(title="Heart Disease and Cholestral",
      xlabel="Age",
      ylabel="Heart Rate")
##4. add legend for ax1
ax1.legend(*scatter.legend_elements(),title="Target");

#5. add a horizontal line for ax1
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

## Add a super title for our figure
fig.suptitle("Heart Disease Data set",fontsize=16,fontweight="bold");
```
